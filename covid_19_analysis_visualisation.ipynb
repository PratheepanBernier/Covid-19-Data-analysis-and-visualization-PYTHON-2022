```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#reading csv file as dataframe
age_df= pd.read_csv("/content/covid-age.csv")
```

```python
#viewing first 5 rows of the dataframe
age_df.head() 
```

```python
#viewing number of rows and columns
age_df.shape 
```

```python
#viewing column names
age_df.columns
```

```python
#dropping rows with null values
age_df.dropna()
```

```python
#using numpy array method for storing column values
age=np.array(age_df["age_classes"])
total_cases=np.array(age_df["total_cases"])
male_cases=np.array(age_df["male_cases"])
female_cases=np.array(age_df["female_cases"])
```

Bar Graph for analysing/visualising relation between Age of the patients and Total positive case:

```python
bar_fig=plt.figure()
bar_axes=bar_fig.add_axes([0,0,1,1])
plt.bar(age,total_cases)
plt.xlabel("Age")
plt.ylabel("Cases")
plt.title("Cases vs Age")
plt.show()
```

Pie chart for visualising distribution of male and female in total cases:

```python
pie_fig=plt.figure()
pie_axes=pie_fig.add_axes([0,0,1,1])
total_male_cases=male_cases.sum()
total_female_cases=female_cases.sum()
Category=["Male","Female"]
Total=[total_male_cases,total_female_cases]
pie_axes.pie(Total,labels=Category,autopct="%1.1f%%")
plt.title("Case vs Gender")
plt.show()
```

Scatter plot for analysing Age and death proportion with respect to Gender:

```python
male_deaths=np.array(age_df["male_deaths"])
female_deaths=np.array(age_df["female_deaths"])
scatter_fig = plt.figure()
scatter_axes = scatter_fig.add_axes([0,0,1,1])
scatter_axes.scatter(age,male_deaths,color="g") #plotting x and sin of x with green colour
scatter_axes.scatter(age,female_deaths,color="r") #plotting x and cos of x with red colour
plt.xlabel("Age group")
plt.ylabel("Death cases")
plt.title("Age vs Death cases\nMale cases:Green\nFemale cases:Red")
plt.show() #viewing
```

Relation Plot for analysing relationship between total positive cases and recovery:

```python
sns.relplot(x="total_cases",y="recovered",data=nation_df)
plt.title("Total cases vs Recovered")
```

Relational line plot for analysing relation between Intensive care and death :

```python
sns.relplot(x="deaths",y="intensive_care",kind="line",data=nation_df)
plt.title("Intensive Care vs Death")
```
